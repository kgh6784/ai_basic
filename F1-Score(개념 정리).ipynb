현재 성능을 평가하는 지표를 dice score만 활용을 하였는데, 다른 추가적인 metric을 사용하고자 생각하다가 `F1-score`를 사용해보기로 했다.

실험의 성능을 확인하기 위해 기본적으로 잘 예측했는지, 예측하지 못했는지 나타내주는 matrix가 있다. 이를 confusion matrix라고 형태는 아래와 같다.

`GT`는 실제 값이다. 쉽게 말해서 실제값과 예측값이 일치하면 True, 실제값과 예측값이 일치하지 않으면 False이다. <br>
우리는 실험 시에 하나만 고려해서는 안 된다. 예를 들어 암의 경우를 살펴보자. 암은 양성(질병 ㅇ)과 음성(질병 x)으로 나뉘는데, 아무리 True Positive가 0이라고 해도 False Negative가 있으면 치명적인 문제가 된다. 이처럼 상황에 맞게 어떤 matrix를 중심적으로 고민하고 사용할지 판단을 할 수 있어야 한다.

이 때 단순 건 수만의 비교가 아닌, 비율을 통해 성능을 평가할 수 있는 방법이 있다. `Precision`과 `Recall`이 있다.
- `precision` : TP / (TP + FP) => 예측한 것 중에 실제 양성 비율
- `recall` : TP / (TP + FN) => 실제 양성 값 중에 잘 예측한 비율

이 때 `precision`과 `recall`을 조화평균한 것이 `F1 Score`이다.
